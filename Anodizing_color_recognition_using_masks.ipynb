Image Processing Library OpenCV is used to process the images. A mask with a threshold having a lower range and having an upper range is defined and a mask is applied to the image. If the color is preset the image shows up in white and else no.


```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
color="red"
```

```python
colors = {'blue': [np.array([95, 255, 85]), np.array([120, 255, 255])],
          'red': [np.array([37, 40, 85]), np.array([178, 255, 255])],
          'yellow': [np.array([16, 0, 99]), np.array([39, 255, 255])],
          'green': [np.array([33, 19, 105]), np.array([77, 255, 255])]}
```

```python
image = cv2.imread("red_ano.png")
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
mask=cv2.inRange(image,colors[color][0],colors[color][1])
```

```python
cv2_imshow(mask)
```

```python
bbox = cv2.boundingRect(mask)
```

```python
print(bbox)
```

```python
if bbox !=(0,0,0,0):
    print("Object detected")
    x, y, w, h = bbox
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
else:
    print("Object not detected")

cv2_imshow(image)
```

```python
hascolor = np.sum(mask)
print(hascolor)
```

```python
if hascolor>0:
  print(f"{color} detected")
else:
  print(f"{color} not detected")
```

```python
res = cv2.bitwise_and(image,image,mask=mask)
fin = np.hstack((image,res))
cv2_imshow(fin)

cv2.waitKey(0)
```
